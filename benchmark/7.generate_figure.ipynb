```python
import plotly.express as px
import pandas as pd
```

Code for generating Figure 2. The data tables from `1.percent_replicating.ipynb` and `4.percent_replicating_dl.ipynb`.

```python
df = pd.concat([pd.read_csv('output/cellprofiler_percent_replicating.csv'), pd.read_csv('output/deepprofiler_percent_replicating.csv')], ignore_index=True, join="inner")

fig = px.bar(data_frame=df,
             x='Perturbation',
             y='Percent_Replicating',
             color='feature_set',
             barmode='group',
             facet_row='time',
             facet_col='Cell')
fig.update_layout(title='Percent Replicating vs. Perturbation',
                  yaxis=dict(title='Percent Replicating'),
                  yaxis3=dict(title='Percent Replicating'))
fig.show("png")
fig.write_image(f'figures/7.figure.png', width=640, height=480, scale=2)
```
