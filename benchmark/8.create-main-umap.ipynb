```python
import pandas as pd
import plotly.express as px
import utils
from umap import UMAP
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

color_discrete_map_seaborn = {
    "Targets": {
        "other targets": "#00000033",
        "CDC25A": "#d55e00cc",
        "LPAR1": "#009e73cc",
        "PLK1": "#cc79a7cc",
        "AURKB": "#0072b2cc",
    }
}

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300

%matplotlib inline
```

UMAP of A549 plates from the primary CPJUMP1 experiment

```python
batch = "2020_11_04_CPJUMP1"
cell = "A549"

experiment_df = (
    pd.read_csv("output/experiment-metadata.tsv", sep="\t")
    .query("Batch==@batch")
    .query("Cell_type==@cell")
    .query("Density==100")
    .query('Antibiotics=="absent"')
)

experiment_df.drop(
    experiment_df[
        (experiment_df.Perturbation == "compound") & (experiment_df.Cell_line == "Cas9")
    ].index,
    inplace=True,
)

genes = [
    "CDC25A",
    "LPAR1",
    "PLK1",
    "AURKB",
]  # common genes with the largest positive cosine similarity across all time points.
```

```python
cosine_sim = (
    pd.read_csv(
        "output/compound_genetic_perturbation_cosine_similarity.csv",
        usecols=[
            "Cell",
            "Metadata_broad_sample",
            "Metadata_matching_target",
            "Genetic_Perturbation",
            "Modality_1_timepoint",
            "Modality_2_timepoint",
            "cosine_sim",
        ],
    )
    .assign(Metadata_gene=lambda x: x.Metadata_matching_target)
    .query("Cell==@cell")
    .query('Genetic_Perturbation=="CRISPR"')
)

metadata = pd.read_csv(
    "../metadata/external_metadata/JUMP-Target-1_compound_metadata.tsv",
    sep="\t",
    usecols=["broad_sample", "pert_iname"],
).rename(columns={"broad_sample": "Metadata_broad_sample"})

cosine_sim_largest = pd.DataFrame()

for timepoints, df in cosine_sim.groupby(
    ["Modality_1_timepoint", "Modality_2_timepoint"]
):
    largest_cosine_sim = (
        df.nlargest(5, "cosine_sim")
        .merge(metadata, on="Metadata_broad_sample")
        .query("Metadata_gene==@genes")
        .query('pert_iname!="danusertib"')
        .rename(
            columns={
                "Modality_1_timepoint": "Compound_timepoint",
                "Modality_2_timepoint": "CRISPR_timepoint",
            }
        )
    )
    cosine_sim_largest = pd.concat(
        [cosine_sim_largest, largest_cosine_sim], ignore_index=True
    )

cosine_sim_largest = (
    cosine_sim_largest.drop_duplicates(
        subset=["Metadata_broad_sample", "Metadata_matching_target"]
    )
)[
    [
        "Metadata_broad_sample",
        "Metadata_matching_target",
        "Metadata_gene",
        "cosine_sim",
        "pert_iname",
    ]
]
```

Read all the plates

```python
all_plates_df = pd.DataFrame()

for modality in experiment_df.Perturbation.unique():
    modality_df = experiment_df.query("Perturbation==@modality")
    for time_point in modality_df.Time.unique():
        time_df = modality_df.query("Time==@time_point")
        for plate in time_df.Assay_Plate_Barcode.unique():
            data_df = (
                utils.load_data(
                    batch, plate, "normalized_feature_select_negcon_batch.csv.gz"
                )
                .assign(Metadata_batch=batch)
                .assign(Metadata_modality=modality)
                .assign(Metadata_cell=cell)
                .assign(Metadata_time=utils.time_point(modality, time_point))
            )
            if modality == "compound":
                data_df.rename(
                    columns={"Metadata_target": "Metadata_gene"}, inplace=True
                )
                data_df = data_df.merge(
                    cosine_sim_largest[
                        [
                            "Metadata_broad_sample",
                            "Metadata_matching_target",
                            "cosine_sim",
                        ]
                    ],
                    on="Metadata_broad_sample",
                    how="left",
                )
            elif modality == "crispr":
                data_df = data_df.merge(
                    cosine_sim_largest[
                        ["Metadata_gene", "Metadata_matching_target", "cosine_sim"]
                    ],
                    on="Metadata_gene",
                    how="left",
                )
            elif modality == "orf":
                data_df["Metadata_matching_target"] = np.nan
                data_df["cosine_sim"] = 0.05
            all_plates_df = utils.concat_profiles(all_plates_df, data_df)

all_plates_df.Metadata_matching_target.fillna("other targets", inplace=True)
all_plates_df.cosine_sim.fillna(0.05, inplace=True)

all_plates_df = utils.remove_empty_wells(all_plates_df)
```

```python
# Use only the long time point
all_plates_df = all_plates_df.query(
    'Metadata_modality!="orf" and Metadata_time=="long"'
)
```

```python
X = utils.get_featuredata(all_plates_df)

umap_2d = UMAP(n_components=2, random_state=157)
projection_2d = umap_2d.fit_transform(X)

batch_list = list(all_plates_df.Metadata_batch)
plate_list = list(all_plates_df.Metadata_Plate)
modality_list = list(all_plates_df.Metadata_modality)
pert_list = np.where(
    all_plates_df.Metadata_pert_type == "control",
    all_plates_df.Metadata_control_type,
    "trt",
)
gene_list = list(all_plates_df.Metadata_gene)
time_list = list(all_plates_df.Metadata_time)
target_list = list(all_plates_df.Metadata_matching_target)
cosine_sim_list = list(all_plates_df.cosine_sim)

X_embedded_df = pd.DataFrame(
    {
        "Batch": batch_list,
        "Plate": plate_list,
        "Modality": modality_list,
        "Pert": pert_list,
        "Gene": gene_list,
        "Time": time_list,
        "Target": target_list,
        "Size": cosine_sim_list,
        "umap 1": projection_2d[:, :1].flatten(),
        "umap 2": projection_2d[:, 1:2].flatten(),
    }
)

X_embedded_df = X_embedded_df.sample(frac=1)
```

```python
# to move the colored points to the top

df1 = X_embedded_df.query("Size==0.05")
df2 = X_embedded_df.query("Size!=0.05")

X_embedded_df = pd.concat([df1, df2], join="inner")
```

```python
g = sns.scatterplot(
    data=X_embedded_df,
    x="umap 1",
    y="umap 2",
    hue=X_embedded_df["Target"],
    style=X_embedded_df["Modality"],
    size=X_embedded_df["Size"],
    sizes=[3, 15, 15, 15, 15],
    hue_order=["other targets", "CDC25A", "LPAR1", "PLK1", "AURKB"],
    palette=color_discrete_map_seaborn["Targets"],
)

h, k = g.get_legend_handles_labels()
del k[6:13]  # remove size label
del h[6:13]  # remove size label
plt.legend(h, k, bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.0, fontsize=10)
g.axes.set_xlabel(g.axes.get_xlabel(), fontsize=15)
g.axes.set_ylabel(g.axes.get_ylabel(), fontsize=15)
sns.despine()
plt.tight_layout()
plt.savefig(f"figures/8.A549_UMAP.png")
```
